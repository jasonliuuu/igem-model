```python
import itertools
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("seaborn")
import pandas as pd
```

```python
FS = (8, 4)  # figure size
```

```python
t = np.linspace(-5, 15, 1000)

fig = plt.figure(figsize=(10, 18))

ax = fig.add_subplot(3, 1, 1)
t0, L = 5., 10000.
for k in [0.5,1.,2.,4.]:
    D = L / (1. + np.exp(-k * (t - t0)))
    _ = plt.plot(t, D, label=f'k={k}')
_ = ax.legend()
_ = ax.group_xlabel('t')

ax = fig.add_subplot(3, 1, 2)
L, k = 10000., 2
for t0 in [2,4,6,8]:
    D = L / (1. + np.exp(-k * (t - t0)))
    _ = plt.plot(t, D, label=f't0={t0}')
_ = ax.legend()
_ = ax.group_xlabel('t')

ax = fig.add_subplot(3, 1, 3)
t0, k = 5., 2
for L in range(4, 8):
    L *= 2000
    D = L / (1. + np.exp(-k * (t - t0)))
    _ = plt.plot(t, D, label=f'L={L}')
_ = ax.legend()
_ = ax.group_xlabel('t')
```

```python
df = pd.read_csv('experimental_results.csv')
```

```python
df
```

```python
total = df.groupby('Time [s]').sum().max(axis=1)
```

```python
all_sets = total.index.to_list()
```

```python
all_sets
```

```python
all_data =  pd.DataFrame()
for set in all_sets:
    temp = df[df['Time [s]'] == set][df.columns[1:]].T.sum(axis=1)
#     temp.index = pd.to_datetime(temp.index)
    temp = temp.to_frame(set)
    all_data = pd.concat([all_data, temp], axis=1)
all_data['avg11'] = all_data[['A01', 'B01','C01']].mean(axis=1)
all_data['avg22'] = all_data[['A02', 'B02','C02']].mean(axis=1)
all_data['avg33'] = all_data[['A03', 'B03','C03']].mean(axis=1)
all_data['avg44'] = all_data[['A04', 'B04','C04']].mean(axis=1)
all_data['avg55'] = all_data[['A05', 'B05','C05']].mean(axis=1)
all_data['avg66'] = all_data[['A06', 'B06','C06']].mean(axis=1)
all_data['avg11_antibio'] = all_data[['A07', 'B07','C07']].mean(axis=1)
all_data['avg22_antibio'] = all_data[['A08', 'B08','C08']].mean(axis=1)
all_data['avg33_antibio'] = all_data[['A09', 'B09','C09']].mean(axis=1)
all_data['avg44_antibio'] = all_data[['A10', 'B10','C10']].mean(axis=1)
all_data['avg55_antibio'] = all_data[['A11', 'B11','C11']].mean(axis=1)
all_data['avg66_antibio'] = all_data[['A12', 'B12','C12']].mean(axis=1)
```

```python
all_data
```

```python
start = 0
ax = all_data[['A01', 'B01','C01','A07','B07','C07','avg11']][start:].plot(style='-', figsize=(15,9))
markers = itertools.cycle(("o", "v", "^", "<", ">", "s", "p", "P", "*", "h", "X", "D", '.'))
for i, line in enumerate(ax.get_lines()):
    marker = next(markers)
    line.set_marker(marker)
_ = ax.legend()
```

```python
# Define funcion with the coefficients to estimate
def logistic(t, a, k, N_m, offset):
    return N_m / (1 + a * np.exp(-k*t)) + offset
```

```python
# Randomly initialize the coefficients
p0 = np.random.random(size=4)
p0
```

```python
bounds = (0., [1000.,3.,1000000.,100.])
```

```python
import scipy.optimize as optim
x = np.array([float(x) for x in all_data.index])
y = np.array([float(x) for x in all_data['avg11']])
```

```python
(a,k,N_m,offset),cov = optim.curve_fit(logistic, x, y, bounds=bounds, p0=p0)
```

```python
a,k,N_m,offset
```

```python
def test_logistic(t):
    return N_m / (1 + a * np.exp(-k*t)) + offset
```

```python
plt.scatter(x, y)
plt.plot(x, test_logistic(x))
plt.title('Logistic Models v. Real Observation of group `avg11`')
plt.legend(['Logistic Model','Experimental Data'])
plt.xlabel('Time/(s)')
plt.ylabel('Concentration')
```

```python
ret_data = {
    'group_name':[],
    'a':[],
    'k':[],
    'N_m':[],
    'offset':[],
    }
```

```python
for group in all_data.columns[all_data.columns.get_loc('avg11'):]:
    cnt += 1
    p0 = np.random.random(size=4)
    y = np.array([float(x) for x in all_data[group]])
    (a,k,N_m,offset),cov = optim.curve_fit(logistic, x, y, bounds=bounds, p0=p0)
    test_logistic = lambda t : N_m / (1 + a * np.exp(-k*t)) + offset
    plt.scatter(x, y)
    ret_data['group_name'].append(group)
    ret_data['a'].append(a)
    ret_data['k'].append(k)
    ret_data['N_m'].append(N_m)
    ret_data['offset'].append(offset)
    plt.plot(x, test_logistic(x),marker = next(markers))
    plt.title(f'Logistic Models v. Real Observation of group {group}')
    plt.legend(['Logistic Model','Experimental Data'])
    plt.xlabel('Time/(s)')
    plt.ylabel('Concentration')
```

```python
output_df = pd.DataFrame(ret_data)
output_df
```
